```python
df = pd.read_excel("/content/Statistics using python_Lancers 10 Nov.xlsx")
df.head()
```

# **Measures of Central tendency**

```python
mean_value = df['Marks'].mean()
median_value = df['Marks'].median()
mode_value = df['Marks'].mode()
```

```python
print("Mean of Marks =", mean_value)
print("Median of Marks =", median_value)
print("Mode of Marks =", mode_value)
```

# **Measures of Central tendency after removing Outlier**

```python
#Check the difference if we remove the outlier
df_no_outlier = df[df['Marks'] < 90]
print("Mean without outlier =", df_no_outlier['Marks'].mean())
print("Median without outlier =", df_no_outlier['Marks'].median())
```

We picked 90 just to remove the obviously extreme 100, and we check min and max because they give us the first clue that an outlier might exist.

# **Measures of Dispersion**

**Minimum and Maximum Marks**

```python
print("Minimum Marks =", df['Marks'].min())
print("Maximum Marks =", df['Marks'].max())
```

**Range**

```python
df['Marks'].max() - df['Marks'].min()
```

**Variance**

```python
df['Marks'].var()
```

**Standard Deviation**

```python
df['Marks'].std()
```

**IQR**

```python
Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
Q1
```

```python
Q3
```

**Detecting outliers using IQR**

```python
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
outliers = df[(df['Marks'] < lower) | (df['Marks'] > upper)]
print("Outliers:\n", outliers)
```

**Standard Deviation and IQR after removing Outlier**

```python
sd_new = df_no_outlier['Marks'].std()
IQR_new = df_no_outlier['Marks'].quantile(0.75) - df_no_outlier['Marks'].quantile(0.25)
print("sd_new:\n", sd_new)
print("IQR_new:\n", IQR_new)
```

Interpretation: overall marks are deviated from mean is +-2.58 and middle 50% of the data is spread or range of 4.0
